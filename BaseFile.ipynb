```python
import pandas as pd
import os
import random
import matplotlib.pyplot as plt
import seaborn as sb
from matplotlib.pyplot import figure
import seaborn as sns
```

```python
baseDir = '/Users/anujjain/Desktop/Side_Project/'
```

```python
train_csv = pd.read_csv(baseDir + 'Loan_project/training_set.csv')
test_csv = pd.read_csv(baseDir + 'Loan_project/testing_set.csv')
```

```python
df = pd.concat([train_csv.assign(ind="train"), test_csv.assign(ind="test")])
```

```python
df.describe()
```

```python
df.info()
```

```python
def fill_na_value(df_fill : pd.DataFrame) -> pd.DataFrame:
    df_fill['Gender'].fillna(random.choice(["Male", "Female"]), inplace=True)
    df_fill['Married'].fillna(random.choice(["No", "Yes"]), inplace=True)
    df_fill['Dependents'].fillna(random.choice(['1','2']), inplace=True)
    df_fill['Education'].fillna('Graduate', inplace=True)
    df_fill['Self_Employed'].fillna(random.choice(["No", "Yes"]), inplace=True)
    df_fill.dropna(subset=['ApplicantIncome'], inplace=True)
    df_fill['CoapplicantIncome'].fillna(df_fill['CoapplicantIncome'].median(), inplace=True)
    df_fill['LoanAmount'].fillna(df_fill['LoanAmount'].mean(), inplace=True)
    df_fill['Loan_Amount_Term'].fillna(df_fill['Loan_Amount_Term'].median(), inplace=True)
    df_fill['Credit_History'].fillna(random.choice([0,1]), inplace=True)    
    return df_fill
fill_na_value(df)
```

```python
for i in df.columns:
    if df[i].isna().any():
        print('####'+ i +'###')
        print(df[i].isnull().sum())
```
