```python
def split_df (df: pd.DataFrame):
    y = df['Loan_Status']
    x = df.drop(['Loan_Status','Loan_ID','ind'], axis = 1 )
    encode = pd.get_dummies(x)
    return y,x,encode
```

```python
def disp_model_report(model,x_train,y_train,x_test,y_test):
    model.fit(x_train,y_train)
    
    tr_pred_lr = model.predict(x_train)
    ts_pred_lr = model.predict(x_test)


    tr_acc_lr = round(accuracy_score(y_train,tr_pred_lr),4)
    ts_acc_lr = round(accuracy_score(y_test,ts_pred_lr),4)
    print("Training Accuracy is:-",round(tr_acc_lr*100,2),"%")
    print("===================================================")
    print("Testing Accuracy is:-",round(ts_acc_lr*100,2),"%")
    print("===================================================")
    print("Classification report for xtest data:-\n\n",classification_report(y_test,ts_pred_lr),"\n")
    print("===================================================")
    print("Confusin Matrix for xtest data:-\n\n")
    cm = confusion_matrix(ts_pred_lr,y_test)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm)
    disp.plot()
    plt.show()
```

```python
def unseen_data(XXnew_test,model) -> pd.DataFrame:
    predicted_probabilities = model.predict_proba(XXnew_test)
    predicted_classes = (predicted_probabilities[:, 1] >= 0.5).astype(int)
    predicted_df = pd.DataFrame(data=XXnew_test, columns=XXnew_test.columns)  # Assuming column names are preserved
    predicted_df['Loan_Status'] = predicted_classes
    return predicted_df
```
